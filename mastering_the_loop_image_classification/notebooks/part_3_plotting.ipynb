```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_baseline = pd.read_csv(r'C:\data\retina_dataset\part_3\baseline_model\000\metrics.csv')
df_baseline['epoch'] = range(len(df_baseline))
```

```python
# sns.set_theme()
# sns.lineplot(data=df, x="train_loss", y='epoch')

sns.set_theme()
f, ax = plt.subplots(1, 1)
# x_col='date'
# y_col = 'count'

ax.plot(df_baseline['epoch'], df_baseline['train_loss'], color="blue", label="Train", linestyle="-")
ax.plot(df_baseline['epoch'], df_baseline["val_loss"], color="red", label="Val", linestyle="-")
# ax.plot_date(df3.date, df3["count"], color="green", label="C", linestyle="-")

ax.legend()
plt.title("Baseline Model")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

## Combining Augmentions with Learning Rate Scheduling

```python
df_aug_step_lr = pd.read_csv(r'C:\data\retina_dataset\part_3\lr_scheduling_plus_augmentation\003\metrics.csv')
df_aug_step_lr['epoch'] = range(len(df_aug_step_lr))
```

```python
# sns.set_theme()
# sns.lineplot(data=df, x="train_loss", y='epoch')

sns.set_theme()
f, ax = plt.subplots(1, 1)
# x_col='date'
# y_col = 'count'

ax.plot(df_aug_step_lr['epoch'], df_aug_step_lr['train_loss'], color="blue", label="Train_Loss", linestyle="-")
ax.plot(df_aug_step_lr['epoch'], df_aug_step_lr["val_loss"], color="red", label="Val_Loss", linestyle="-")
ax.plot(df_aug_step_lr['epoch'], df_aug_step_lr['train_accuracy'], color="orange", label="Train_Accuracy", linestyle="-")
ax.plot(df_aug_step_lr['epoch'], df_aug_step_lr["val_accuracy"], color="yellow", label="Val", linestyle="-")
# ax.plot_date(df3.date, df3["count"], color="green", label="C", linestyle="-")

ax.legend(loc='best')
plt.title("Baseline Model")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
# sns.set_theme()
# sns.lineplot(data=df, x="train_loss", y='epoch')

sns.set_theme()
f, ax = plt.subplots(1, 1)
# x_col='date'
# y_col = 'count'

ax.plot(df_aug_step_lr['epoch'], df_aug_step_lr['learning_rate'], color="blue", label="LearningRate", linestyle="-")
# ax.plot_date(df3.date, df3["count"], color="green", label="C", linestyle="-")

ax.legend(loc='best')
plt.title("Baseline Model")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```
